Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов.
Даны значения роста в трех группах случайно выбранных спортсменов:

Футболисты: 173, 175, 180, 178, 177, 185, 183, 182.

Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180.

Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

```python
import numpy as np
from scipy import stats
```

```python
x1=np.array([173, 175, 180, 178, 177, 185, 183, 182])
x2=np.array([177, 179, 180, 188, 177, 172, 171, 184, 180])
x3=np.array([172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170])
```

```python
f=stats.f_oneway(x1,x2,x3)
f
```

H0 - нет различий среднего роста среди взрослых футболистов, хоккеистов и штангистов (H0: μ1=μ2=μ3);<br>
H1 - разлия есть.
1. *pvalue=0.01048* < a=0.05 - на уровне статистической значимости **a=0.05** отвергаем нулевую гипотезу. И принимаем альтернативную, о том что: значения роста в трех группах случайно выбранных спортсменов **различны**.
2. При этом на уровне статистической значимости **a=0.01** < *pvalue=0.01048* - нулевую гипотезу мы отвергнуть не можем. Следовательно оставляем нулевую: значения роста в трех группах случайно выбранных спортсменов **НЕ** различны.


```python
import matplotlib.pyplot as plt
# plt.scatter()
plt.plot(x1, c='g', label='Футболисты')
plt.plot(x2, c='y', label='Хоккеисты')
plt.plot(x3, c='r', label='Штангисты')
plt.legend()
plt.ylabel('Рост', rotation=90)
plt.show()
```
