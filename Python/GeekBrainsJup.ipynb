```python
from sympy import *
from sympy.plotting import plot
init_printing()
```

```python
import math
```

```python
x = Symbol('x')
```

```python
f = -12*x**4 - 18*x**3 + 5*x**2 + 10*x - 30
f
```

```python
plot(f)
plot(diff(f))
print(f)
print(diff(f))
```

```python
x1, x2, x3, x4 = solve(f)
```

```python
z1 = round(x1.evalf(), 2)
z2 = round(x2.evalf(), 2)
z3 = round(x3.evalf(), 2)
z4 = round(x4.evalf(), 2)
```

```python
print([z1, z2, z3, z4])
```
